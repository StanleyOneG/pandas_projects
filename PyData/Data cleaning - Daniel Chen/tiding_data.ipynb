```python
weather = pd.read_csv('./pydatadc_2018-tidy-master/data/weather.csv')
```

```python
weather.sample(10)
```

```python
weather_melt = pd.melt(
    weather,
    id_vars = ['id', 'year', 'month', 'element'],
    var_name = 'day',
    value_name = 'temp'
)
```

```python
weather_melt.head()
```

```python
# 'id_vars' in melt() - what we don't want to 'touch'
# 'index' in pivot_table() - what we don't want to 'touch'
# 'columns' in pivot_table() - what we want to 'pivot' or values that we want to make columns from

(weather
 .melt(id_vars = ['id', 'year', 'month', 'element'], 
       var_name = 'day', 
       value_name = 'temp')
 .pivot_table(index = ['id', 'year', 'month', 'day'],
              columns = 'element',
              values = 'temp')
 .reset_index()
 # .dtypes
 # .memory_usage(deep=True)
 # .sum()
)
```
